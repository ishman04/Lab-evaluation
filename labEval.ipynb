```python
import pandas as pd
from google.colab import files
```

```python
print("Please upload the dataset CSV file:")
uploaded = files.upload()
```

```python
file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)
```

```python
target_column = 'Target3'
df = df.drop(columns=['Target1', 'Target2', 'Target4', 'Target5'], errors='ignore')
```

```python
exp = setup(data=df, target=target_column, train_size=0.7, session_id=42, verbose=False)
```

```python
best_model = compare_models()
```

```python
print("\nBest Model Identified:")
print(best_model)
```

```python
save_model(best_model, 'best_regression_model')
```

```python
predictions = predict_model(best_model)
print(predictions)
```
